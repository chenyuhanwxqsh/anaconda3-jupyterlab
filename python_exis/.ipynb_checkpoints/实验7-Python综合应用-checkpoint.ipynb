1. 下载一篇格式为文本文件的中文小说，制作高频词云并统计出场次数最多的10个人物。

```python
import jieba
import wordcloud

f = open("test.txt","rb")
t = f.read()
f.close()
ls = jieba.lcut(t)
txt = " ".join(ls)
w = wordcloud.WordCloud( font_path = "NotoSerifCJK-Bold.ttc",
width = 1000,height = 700,background_color = "white",
)

w.generate(txt)
w.to_file("test_1.png")
```

```python
import jieba
excludes = {"来到","人马","领兵","将军","却说","荆州","二人","不可","不能","如此"}
txt = open("test.txt", "rb").read()
words = jieba.lcut(txt)
counts = {}
for word in words:
if len(word) == 1:
continue
elif word == "诸葛亮" or word == "孔明曰":
rword = "孔明"
elif word == "关公" or word == "云长":
rword = "关羽"
elif word == "玄德" or word == "玄德曰":
rword = "刘备"
elif word == "孟德" or word == "丞相":
rword = "曹操"
else:
rword = word
counts[rword] = counts.get(rword,0) + 1
for word in excludes:
del(counts[word])
items = list(counts.items())
items.sort(key=lambda x:x[1], reverse=True)
for i in range(55):
word, count = items[i]
print ("{0:<10}{1:>5}".format(word, count))
```

2. 完成教材216页的课后练习第1题。

```python
import numpy as np
ls=np.array([1,0,2,0,0,1])
f=np.poly1d(ls)
fNum=f([2,5])
print(f)
print(fNum)
```

```python
fDer1=np.polyder(f,1)
fDer2=np.polyder(f,2)
print(fDer1)
print(fDer2)
```

3. 完成教材216页的课后练习第2题。

```python
import numpy as np
import matplotlib.pyplot as plt
x1=np.linspace(0,2*np.pi,100,endpoint=True)
x2=np.linspace(-2,2,100,endpoint=True)
y1=np.sin(x1)+x1**2
y2=x2**3+2*x2**2+1
plt.plot(x1,y1,'r',x2,y2,'g')
plt.legend(['$y=sin(x)+x^2$','$y=x^3+2*x^2+1$'])
plt.show()
```
