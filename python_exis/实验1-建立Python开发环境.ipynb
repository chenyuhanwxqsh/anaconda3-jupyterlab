(1)在安装Jupter Lab前将通道设为国内的conda和pypi镜像源

```python
ssl_verify: true
channels:
  - https://mirrors.tuna.tsinghua.edu.cn/anaconda/pkgs/free/
  - defaults
show_channel_urls: true
```

.condarc配置文件在C:\Users\Dell\.condarc中

* pypi通道：

```python
pip config set global.index-url https://pypi.tuna.tsinghua.edu.cn/simple
```

会将配置信息写入C:\Users\yun\AppData\Roaming\pip\pip.ini文件中，也可手动创建
文件C:\Users\yun\pip\pip.ini文件并写入配置信息，但该文件优先级低于前面的。
